```python
#load tools

from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import homogeneity_completeness_v_measure

import numpy as np
import matplotlib.pyplot as plt
```

```python
#load data
datasets = load_iris()

X,y = datasets.data,datasets.target
print(datasets.target_names)
```

```python
#training
clus = KMeans(n_clusters=3)
y_predict = clus.fit_predict(X)
```

```python
#evaluate tools
h,c,v = homogeneity_completeness_v_measure(y,y_predict)
print(f'Homogeneity: {h}')
print(f'completeness: {c}')
print(f'v-measure: {v}')
```

```python
#transpose data for plot scatter 

#print(X)
print(datasets.feature_names)
X_t = np.transpose(X)
#print(X_t[0])
#print(X_t[1])
```

```python
plt.subplot(1,2,1)
plt.scatter(X_t[0],X_t[1],c=y)
plt.subplot(1,2,2)
plt.scatter(X_t[0],X_t[1],c=y_predict)
print(datasets.target_names)
```
