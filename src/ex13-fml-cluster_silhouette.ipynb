```python
#import tools
from sklearn.datasets import make_blobs
from sklearn.cluster import  KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Create datas
X,y = make_blobs(n_samples=100,n_features=2,center_box= (0,10),centers=5,cluster_std=1)

print(X.shape)
X_t = np.transpose(X)
print(X_t.shape)
plt.scatter(X_t[0],X_t[1])
```

```python
#training models
clus = KMeans(n_clusters=3)
y_predict = clus.fit_predict(X)
```

```python
#evaluate tools
print(f'Silhouette_score: {silhouette_score(X,y_predict)}')
```

```python
plt.subplot(1,2,1)
plt.scatter(X_t[0],X_t[1])
plt.subplot(1,2,2)
plt.scatter(X_t[0],X_t[1],c=y_predict)
```
